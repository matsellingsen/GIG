```python
from optimum.intel.openvino import OVModelForCausalLM
from transformers import AutoTokenizer

MODEL_DIR = r"C:\Users\matse\gig\models\qwen2.5-1.5b-instruct"
OUTPUT_DIR = r"C:\Users\matse\gig\models\qwen2.5-1.5b-instruct-ir4"

SEQ_LEN = 256  # choose 128/256/512/1024 depending on your needs

# Load tokenizer locally
tokenizer = AutoTokenizer.from_pretrained(
    MODEL_DIR,
    local_files_only=True
)

# Force static shapes for ALL inputs
model = OVModelForCausalLM.from_pretrained(
    MODEL_DIR,
    export=True,
    compile=False,
    local_files_only=True,
    ov_config={"INFERENCE_PRECISION_HINT": "int8"},
    input_shapes={
        "input_ids": (1, SEQ_LEN),
        "attention_mask": (1, SEQ_LEN),
        "position_ids": (1, SEQ_LEN),
        "beam_idx": (1,),  # required by the exporter
    }
)

model.save_pretrained(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)

print("Export complete. IR saved to:", OUTPUT_DIR)
```
