```python
from optimum.intel.openvino import OVModelForCausalLM
from transformers import AutoTokenizer

# Path to your local Qwen2.5-1.5B-Instruct folder
MODEL_DIR = "C:\\Users\\matse\\gig\\models\\qwen2.5-1.5b-instruct"          # <-- change this to your local folder
OUTPUT_DIR = "C:\\Users\\matse\\gig\\models\\qwen2.5-1.5b-instruct-ir"   # <-- where IR files will be written

# Load tokenizer fully offline
tokenizer = AutoTokenizer.from_pretrained(
    MODEL_DIR,
    local_files_only=True
)

# Export to OpenVINO IR (INT8 + static shapes)
model = OVModelForCausalLM.from_pretrained(
    MODEL_DIR,
    export=True,
    compile=False,
    load_in_8bit=True,          # INT8 quantization
    static_shapes=True,         # required for Intel NPU
    local_files_only=True
)

# Save IR model + tokenizer
model.save_pretrained(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)

print("Export complete. IR saved to:", OUTPUT_DIR)
```
