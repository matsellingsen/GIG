```python
#load json
import json

with open("C:\\Users\\matse\\gig\\src\\system_v5\\tests\\dataset\\new_synthetic_labelled_input.json", "r", encoding="utf-8") as f:
    json_dict = json.load(f)

# Count for each main.key, the number of items its keys has.
total = 0
for item in json_dict.keys():
    main_key = 0
    for i, key in enumerate(json_dict[item]):
        main_key += len(json_dict[item][key])
    print(f"{item} has {main_key} items")
    total += main_key
print(total)
```
